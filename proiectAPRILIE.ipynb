# Descrierea setului de date

<h2>Ce știm din seturile noastre de date:</h2>

1.   [transport aerian de mărfuri](https://ec.europa.eu/eurostat/databrowser/view/ttr00011/default/table?lang=en&category=t_avia) ⟶ numărul total de **tone** transportate *anual*
2.   [transport aerian de pasageri](https://ec.europa.eu/eurostat/databrowser/view/ttr00016/default/table?lang=en&category=t_avia) ⟶ numărul total de **pasageri** transportați *lunar*
3.   [transport aerian de pasageri](https://ec.europa.eu/eurostat/databrowser/view/ttr00012/default/table?lang=en&category=t_avia) ⟶ numărul total de **pasageri** transportați *anual*

Din cele 3 seturi de date menționate, s-au realizat niște **măsurători de transport aerian** pentru pasageri & marfă & statistici regionale.

<h2>Ce putem să observăm din seturile de date și ce reprezintă de fapt?</h2>

La prima coloană din orice set de date specificat, se poate găsi o structură generalizată:

1. <center><h1>freq,unit,tra_meas,schedule,tra_cov,geo\TIME_PERIOD</h1></center>

2. <center><h1>freq,unit,schedule,tra_cov,tra_meas,geo\TIME_PERIOD</h1></center>

3. <center><h1>freq,unit,tra_meas,tra_cov,schedule,geo\TIME_PERIOD</h1></center>

Fiecare câmp are o semnificație a datelor ⇔ ce reprezintă fiecare valoare.

---


În cazul nostru, avem:

*   `freq` ⟶ frecvență
    *  `M` ⟶ lunar
    *  `A` ⟶ anual
*   `unit` ⟶ unitatea de măsură
    *  `PAS` ⟶ numărul total de pasageri
    *  `T` ⟶ numărul total de tone
*   `schedule` ⟶ tipul exact de zbor
    *  `TOT` ⟶ orice tip de zbor
    *  `SCH` ⟶ zbor regulat - pasageri
    *  `NSC` ⟶ zbor neregulat - privat
*   `tra_meas` ⟶ transport measurement - indicator de transport
    *  `PAS_CRD` ⟶ numărul total de pasageri transportați

    [CRD = carried]

    *  `FRM_LD_NLD` ⟶ numărul total de marfă transportate

    [FRM = marfă - tip, LD = loaded - acțiune, NLD = total - acoperire]

    Alte exemple pot fi:

    *  `PAS_EMB` ⟶ numărul total de pasageri imbarcați
    *  `PAS_DIS` ⟶ numărul total de pasageri debarcați

    [EMB/DIS = embarked/disembarked passengers]

*   `tra_cov` ⟶ transport coverage - zonă de transport
    *  `NAT` ⟶ național
    *  `INTL` ⟶ internațional   
    *  `TOTAL` ⟶ global
*   `geo` ⟶ denumirea țării - de exemplu: RO = România
*   `TIME_PERIOD` ⟶ perioadă de timp

---

<h2>Perioadele sunt setate în felul următor:</h2>
