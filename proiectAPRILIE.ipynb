# Descrierea setului de date

<h2>Ce știm din seturile noastre de date?</h2>

La prima coloană din orice set de date specificat, se poate găsi o structură generalizată.

1.   [transport aerian de mărfuri](https://ec.europa.eu/eurostat/databrowser/view/ttr00011/default/table?lang=en&category=t_avia) ⟶ numărul total de **tone** transportate *anual*
  <center><h1>freq,unit,tra_meas,schedule,tra_cov,geo\TIME_PERIOD</h1></center>

2.   [transport aerian de pasageri](https://ec.europa.eu/eurostat/databrowser/view/ttr00016/default/table?lang=en&category=t_avia) ⟶ numărul total de **pasageri** transportați *lunar*
<center><h1>freq,unit,schedule,tra_cov,tra_meas,geo\TIME_PERIOD</h1></center>

3.   [transport aerian de pasageri](https://ec.europa.eu/eurostat/databrowser/view/ttr00012/default/table?lang=en&category=t_avia) ⟶ numărul total de **pasageri** transportați *anual*
<center><h1>freq,unit,tra_meas,tra_cov,schedule,geo\TIME_PERIOD</h1></center>

Din cele 3 seturi de date menționate, s-au realizat niște **măsurători de transport aerian** pentru pasageri & marfă & statistici regionale.

Fiecare câmp are o semnificație a datelor ⇔ ce reprezintă fiecare valoare.


---


<h2>Ce putem să observăm din seturile de date și ce reprezintă de fapt?

În cazul nostru, avem:

*   `freq` ⟶ frecvență
    *  `M` ⟶ lunar
    *  `A` ⟶ anual
*   `unit` ⟶ unitatea de măsură
    *  `PAS` ⟶ numărul total de pasageri
    *  `T` ⟶ numărul total de tone
*   `schedule` ⟶ tipul exact de zbor
    *  `TOT` ⟶ orice tip de zbor
    *  `SCH` ⟶ zbor regulat - pasageri
    *  `NSC` ⟶ zbor neregulat - privat
*   `tra_meas` ⟶ transport measurement - indicator de transport
    *  `PAS_CRD` ⟶ numărul total de pasageri transportați

    [CRD = carried]

    *  `FRM_LD_NLD` ⟶ numărul total de marfă transportate

    [FRM = marfă - tip, LD = loaded - acțiune, NLD = total - acoperire]

    Alte exemple pot fi:

    *  `PAS_EMB` ⟶ numărul total de pasageri imbarcați
    *  `PAS_DIS` ⟶ numărul total de pasageri debarcați

    [EMB/DIS = embarked/disembarked passengers]

*   `tra_cov` ⟶ transport coverage - zonă de transport
    *  `NAT` ⟶ național
    *  `INTL` ⟶ internațional   
    *  `TOTAL` ⟶ global
*   `geo` ⟶ denumirea țării - de exemplu: RO = România
*   `TIME_PERIOD` ⟶ perioadă de timp

# Interpretarea numerelor, graficelor și concluziile relevante proiectului

Putem să prezentăm mai multe concluzii generale care sunt formulate pe baza calculelor matematice și reprezentărilor grafice:

*   din punct de vedere al transportului de mărfuri
    *   țările cu cantități semnificative de mărfuri transportate gen Luxemburg sau Belgia nu sunt influențate din cauză că nivelul de comerț [cumpărături online & necesități medicale] este crescut.
    *   majoritatea țărilor au obținut rezultatele scăzute deoarece totul depinde de situația lor în politică, mai exact, ce priorități are fiecare țară și importanța ei în UE la transport de mărfuri.
    *   impactul rămâne ridicat. Practic, transportul de colete au crescut enorm datorită comenzilor online si e-commerce-ul care "s-a viralizat".
*   din punct de vedere al transportului de pasageri
    *   fiecare țară din Uniunea Europeană este afectată din cauza măsurilor restrictive - lockdown, pașaport COVID, teste PCR la intrare în anumite țări, distanțare fizică obligatorie, ș.a.m.d.
    *   în momentul în care au apărut mai multe soluții impotriva COVID-ului gen vaccinare obligatorie, teste PCR și respectarea măsurilor sanitare enumerate publicului, putem considera că transportul de pasageri a început să crească ușor-ușor treptat.

Dacă ne uităm la rezultatele obținute totale pentru cele 27 de țări din Uniunea Europeană, putem să tragem o concluzie finală că pandemia COVID a încetinit transportul aerian de pasageri, dar și nu transportul aerian de marfă.
